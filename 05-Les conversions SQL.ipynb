# Extraire des informations
## EXTRACT

```python
%%sql 
SELECT 
    NOM, 
    EXTRACT (YEAR FROM DATE_EMBAUCHE)  "Année",
    EXTRACT (MONTH FROM DATE_EMBAUCHE) "Mois",
    EXTRACT (DAY FROM DATE_EMBAUCHE)   "Jour"
FROM EMPLOYES ORDER BY 2,3,4
FETCH FIRST 5 ROWS ONLY
```

# Chaîne vers autres types
## TO_NUMBER

```python
%%sql 
SELECT
    TO_NUMBER('-1234567890','S9999999999') "Colonne 1",
    TO_NUMBER('+1234567890','S9999999999') "Colonne 2",
    TO_NUMBER('+123.456'   ,'S999.999'   ) "Colonne 3",
    TO_NUMBER('-123.456'   ,'999.999'    ) "Colonne 4"
FROM DUAL
```

## TO_DATE

```python
%%sql 
SELECT TO_DATE( '10/04/2021','DD/MM/YYYY'), 
       TO_DATE( '12/10/2021 20:12:00', 
                'DD/MM/YYYY HH24:MI:SS') "TO_DATE" 
FROM DUAL
```

# Les fonctions générales
## CASE

```python
%%sql 
SELECT 
    NOM||' '||PRENOM "Employé", 
    FONCTION,
    CASE FONCTION
        WHEN 'Président'  THEN
             SALAIRE*1.05
        WHEN 'Vice-Président'  THEN
             SALAIRE*1.1
        WHEN 'Chef des ventes' THEN
             SALAIRE*1.2
        WHEN 'Représentant(e)' THEN
             SALAIRE*1.1 + COMMISSION
        ELSE
             SALAIRE
    END "Salaire"
FROM EMPLOYES
ORDER BY SALAIRE DESC 
FETCH FIRST 15 ROWS ONLY
```

```python
%%sql 
SELECT 
    NOM, 
    FONCTION, 
    SALAIRE,
    CASE 
        WHEN FONCTION = 'Assistante commerciale'
        THEN '10%'
        WHEN FONCTION = 'Représentant(e)' AND
          SALAIRE between 6000 AND 6400 
        THEN '30%'
        WHEN FONCTION = 'Représentant(e)' AND
          SALAIRE < 6000
        THEN '20%'
        ELSE
          'Pas d''augmentation'
    END "Salaire" 
FROM EMPLOYES 
ORDER BY SALAIRE DESC 
OFFSET 80 ROWS FETCH NEXT 25 ROWS ONLY
```

## NULLIF

```python
%%sql 
SELECT 
    COMMISSION, 
    NULLIF(COMMISSION, 1000) RETOUR1,
    NOM||' '||PRENOM "Employé", 
    NULLIF(LENGTH(NOM),
    LENGTH(PRENOM)) RETOUR2 
FROM EMPLOYES
FETCH FIRST 10 ROWS ONLY
```

# COALESCE

```python
%%sql 
SELECT 
    NOM_PRODUIT,
    UNITES_COMMANDEES, 
    UNITES_STOCK, 
    INDISPONIBLE,
        CASE
            WHEN UNITES_COMMANDEES IS NOT NULL
            THEN 'Unités commandées = '
            WHEN UNITES_STOCK IS NOT NULL
            THEN 'Unités en stock = '
            ELSE
                'Produit indisponible = '
        END 
    || COALESCE(UNITES_COMMANDEES, UNITES_STOCK, INDISPONIBLE) "Stock des produits"
FROM PRODUITS
```
