# La clause de partitionnement

<img src="https://raw.githubusercontent.com/rbizoi/IntelligenceEnDonneesDeSante/main/InterrogationSQL/images/M06-10.png" width="400">


```python
%%sql 
SELECT NO_EMPLOYE, FONCTION, SALAIRE,
SUM( SALAIRE) OVER ( PARTITION BY FONCTION) "Somme Salaire"
FROM EMPLOYES
```

```python
%%sql
SELECT 
    NOM_PRODUIT "Produit", 
    CODE_CATEGORIE "Cat",
    NO_FOURNISSEUR "Four", 
    UNITES_STOCK "US",
    SUM(UNITES_STOCK) OVER (PARTITION BY CODE_CATEGORIE) "SUSC",
    AVG(UNITES_STOCK) OVER (PARTITION BY NO_FOURNISSEUR) "AUSF"
FROM PRODUITS
```

```python
%%sql
SELECT 
    NOM, 
    FONCTION, SALAIRE, 
    SALAIRE * 100/ SUM( SALAIRE) OVER ( PARTITION BY FONCTION) "% Fonction",
    SALAIRE * 100/ SUM( SALAIRE) OVER () "% Total" 
FROM EMPLOYES
```

```python
%%sql
SELECT 
    PAYS, 
    ANNEE, 
    TRIMESTRE T, 
    SUM(PRIX_UNITAIRE*QUANTITE) CA,
    SUM(SUM(PRIX_UNITAIRE*QUANTITE)) OVER( PARTITION BY PAYS) "S Pays",
    SUM(SUM(PRIX_UNITAIRE*QUANTITE)) OVER( PARTITION BY ANNEE) "S Année"
FROM CLIENTS CL                                
     JOIN COMMANDES CO                         
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)  
     JOIN DETAILS_COMMANDES DC                 
        ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)
GROUP BY PAYS, ANNEE, TRIMESTRE 
```

# La clause d’ordre

<img src="https://raw.githubusercontent.com/rbizoi/IntelligenceEnDonneesDeSante/main/InterrogationSQL/images/M06-11.png" width="400">


```python
%%sql 
SELECT 
    CL.SOCIETE, 
    CO.DATE_COMMANDE, 
    CO.PORT P,
    SUM(CO.PORT) OVER ( PARTITION BY CO.CODE_CLIENT ORDER BY CO.DATE_COMMANDE) SCC,
    SUM(CO.PORT) OVER ( ORDER BY CO.CODE_CLIENT, CO.DATE_COMMANDE)  SCT
FROM CLIENTS CL                                
     JOIN COMMANDES CO                         
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)        
WHERE ANNEE = 2011 AND MOIS = 5 AND PAYS = 'France'
```

# La clause de fenêtrage

<img src="https://raw.githubusercontent.com/rbizoi/IntelligenceEnDonneesDeSante/main/InterrogationSQL/images/M06-12.png" width="400">


```python
%%sql 
SELECT 
    CL.SOCIETE, 
    CO.DATE_COMMANDE, 
    CO.PORT P,
    SUM(CO.PORT) OVER ( PARTITION BY CO.CODE_CLIENT 
                        ORDER BY CO.DATE_COMMANDE
                        ROWS BETWEEN 1 PRECEDING AND 1 PRECEDING )  C01,
    SUM(CO.PORT) OVER ( PARTITION BY CO.CODE_CLIENT 
                        ORDER BY CO.DATE_COMMANDE
                        ROWS BETWEEN 1 FOLLOWING AND 1 FOLLOWING )  C02
FROM CLIENTS CL                                
     JOIN COMMANDES CO                         
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)        
WHERE ANNEE = 2011 AND MOIS = 5 AND PAYS = 'France'    
```

```python
%%sql 
SELECT 
    CL.SOCIETE, 
    CO.DATE_COMMANDE, 
    CO.PORT P,
    SUM(CO.PORT) OVER ( ORDER BY CO.CODE_CLIENT, CO.DATE_COMMANDE)  SCT,
    SUM(CO.PORT) OVER ( ORDER BY CO.CODE_CLIENT, CO.DATE_COMMANDE
                        ROWS UNBOUNDED PRECEDING                         )  C01,    
    SUM(CO.PORT) OVER ( ORDER BY CO.CODE_CLIENT, CO.DATE_COMMANDE
                        ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW )  C02
FROM CLIENTS CL                                
     JOIN COMMANDES CO                         
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)        
WHERE ANNEE = 2011 AND MOIS = 5 AND PAYS = 'France'       
FETCH NEXT 18 ROWS WITH TIES
```

```python
%%sql 
SELECT 
DATE_COMMANDE, 
SUM(PORT) SP, 
AVG(SUM(PORT)) OVER ( ORDER BY DATE_COMMANDE 
                      ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING ) AP
FROM CLIENTS CL                                
     JOIN COMMANDES CO                         
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT) 
WHERE ANNEE = 2011 AND MOIS = 5 
GROUP BY DATE_COMMANDE
FETCH NEXT 18 ROWS WITH TIES
```

# Fonctions de classement

## ROW_NUMBER

```python
%%sql 
SELECT 
    CODE_CATEGORIE CAT,
    NO_FOURNISSEUR NF,
    UNITES_STOCK US,
    ROW_NUMBER() OVER ( PARTITION BY CODE_CATEGORIE
                       ORDER BY UNITES_STOCK)       RN1,
    ROW_NUMBER() OVER ( PARTITION BY CODE_CATEGORIE
                 ORDER BY UNITES_STOCK NULLS FIRST) RN2,
    ROW_NUMBER()OVER (
       ORDER BY CODE_CATEGORIE, UNITES_STOCK)       RN3,
    ROW_NUMBER() OVER ( ORDER BY CODE_CATEGORIE,
                 NO_FOURNISSEUR,UNITES_STOCK)       RN4,
    ROW_NUMBER() OVER ( ORDER BY NO_FOURNISSEUR,
                           UNITES_STOCK  )          RN5
FROM PRODUITS WHERE CODE_CATEGORIE < 3
```

## RANK

```python
%%sql 
SELECT 
    ROW_NUMBER() OVER ( ORDER BY SALAIRE DESC) RN, 
    NOM,
    SALAIRE,
    RSD,
    RSC
FROM (SELECT 
          NOM,
          SALAIRE, 
          RANK ()OVER ( ORDER BY SALAIRE DESC) RSD,
          RANK ()OVER ( ORDER BY SALAIRE ASC ) RSC
      FROM EMPLOYES) 
WHERE RSD <= 3 or RSC <= 3
```

## DENSE_RANK

```python
%%sql
SELECT 
    ROW_NUMBER() OVER 
            ( ORDER BY UNITES_STOCK NULLS LAST) RN,
    NOM_PRODUIT, NO_FOURNISSEUR NF, UNITES_STOCK US,
    RANK() OVER ( ORDER BY UNITES_STOCK) R,
    DENSE_RANK() OVER (ORDER BY UNITES_STOCK) DR,
    DENSE_RANK() OVER (ORDER BY UNITES_STOCK
                      NULLS FIRST) DRI
FROM PRODUITS 
WHERE CODE_CATEGORIE = 1
ORDER BY US NULLS LAST
```

## PERCENT_RANK

```python
%%sql 
SELECT 
    FONCTION, 
    PAYS, 
    SUM(SALAIRE) SS,
    PERCENT_RANK() OVER ( PARTITION BY FONCTION
           ORDER BY SUM(SALAIRE)) PERCENT_RANK
FROM EMPLOYES 
GROUP BY FONCTION, PAYS
```

## NTILE

```python
%%sql 
SELECT 
    ROW_NUMBER() OVER(ORDER BY SUM(PRIX_UNITAIRE*QUANTITE)) RN,
    PAYS, 
    SUM(PRIX_UNITAIRE*QUANTITE) CA,
    NTILE(4) OVER(ORDER BY SUM(PRIX_UNITAIRE*QUANTITE)) NT
FROM CLIENTS CL                                
    JOIN COMMANDES CO                         
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)  
    JOIN DETAILS_COMMANDES DC                 
        ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)
WHERE ANNEE = 2011 
GROUP BY PAYS
```

## WIDTH_BUCKET

```python
%%sql
SELECT 
    PAYS, 
    SUM(QUANTITE) Q,
    ROUND(MAX(SUM(QUANTITE))OVER()   / 4) B1,
    ROUND(MAX(SUM(QUANTITE))OVER()   / 2) B2,
    ROUND(MAX(SUM(QUANTITE))OVER()*3 / 4) B3,
    NTILE(4) OVER(ORDER BY SUM(QUANTITE)) NT,
    WIDTH_BUCKET( SUM(QUANTITE), 
                  MIN(SUM(QUANTITE)) OVER (), 
                  MAX(SUM(QUANTITE)) OVER ()+1, 
                  4 ) WB
FROM CLIENTS CL                                
     JOIN COMMANDES CO                         
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)  
     JOIN DETAILS_COMMANDES DC                 
        ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)
WHERE ANNEE = 2011 
GROUP BY PAYS
```

# Fonctions de fenêtre
## FIRST_VALUE

```python
%%sql 
WITH VENTES AS (
SELECT CL.PAYS,CO.DATE_COMMANDE,SUM(CO.PORT) SP
    FROM CLIENTS CL JOIN COMMANDES CO
         ON ( CL.CODE_CLIENT = CO.CODE_CLIENT
             AND CO.ANNEE = 2011 AND CO.MOIS = 5
             AND CL.PAYS IN ('Suisse'))
GROUP BY CL.PAYS,CO.DATE_COMMANDE)
SELECT PAYS,JOUR,SP,
       FIRST_VALUE( SP) OVER
         ( ORDER BY JOUR RANGE INTERVAL '1' DAY PRECEDING) FV1,
       FIRST_VALUE( SP) OVER ( ORDER BY JOUR
                               RANGE 1 PRECEDING )         FV2,
       FIRST_VALUE( SP) OVER ( ORDER BY JOUR)              FV3,
       FIRST_VALUE( SP) IGNORE NULLS OVER ( ORDER BY JOUR) FV4
FROM VENTES V
     PARTITION BY (PAYS) RIGHT OUTER JOIN
      ( SELECT JOUR FROM DIM_TEMPS
        WHERE  ANNEE = 2011 AND MOIS_N = 5) T
     ON ( V.DATE_COMMANDE = T.JOUR)
ORDER BY PAYS,JOUR
FETCH NEXT 10 ROWS WITH TIES
```

## LAST_VALUE

```python
%%sql
SELECT PAYS, MOIS, SUM(DC.QUANTITE) Q,
       LAST_VALUE( SUM(DC.QUANTITE)) OVER
         ( PARTITION BY PAYS ORDER BY MOIS ROWS
                  BETWEEN CURRENT ROW AND 1 FOLLOWING ) LV1,
       LAST_VALUE( SUM(DC.QUANTITE)) OVER
                    ( PARTITION BY PAYS ORDER BY MOIS ) LV2,
       LAST_VALUE( SUM(DC.QUANTITE)) OVER
                    ( PARTITION BY PAYS               ) LV3
FROM CLIENTS CL                                
     JOIN COMMANDES CO                         
        ON ( CL.CODE_CLIENT = CO.CODE_CLIENT)  
     JOIN DETAILS_COMMANDES DC                 
        ON ( CO.NO_COMMANDE = DC.NO_COMMANDE)  
     JOIN PRODUITS PR                          
        ON ( DC.REF_PRODUIT = PR.REF_PRODUIT)
WHERE ANNEE = 2010 AND PAYS IN ('France','Suisse')
GROUP BY PAYS, MOIS
```
