# Opérateurs arithmétiques

Une expression arithmétique est une combinaison de noms de colonnes, de constantes et de fonctions arithmétiques (les fonctions arithmétiques sont traitées plus loin) combinées au moyen des opérateurs arithmétiques addition « + », soustraction « - », multiplication « * » ou division « / ».


```python
%%sql 
SELECT 
    NOM_PRODUIT "Produit", 
    PRIX_UNITAIRE*UNITES_STOCK "Stock",
    UNITES_COMMANDEES*12 "Commandes" 
FROM PRODUITS
FETCH NEXT 5 ROWS ONLY
```

```python
%%sql 
SELECT 
    NOM||' '||PRENOM as "Employé", 
    SALAIRE * 1.1 as "Salaire"
FROM EMPLOYES
FETCH NEXT 5 ROWS ONLY
```

```python
%%sql 
SELECT 
    NOM,
    DATE_NAISSANCE,
    DATE_NAISSANCE+1,
    (SYSDATE-DATE_NAISSANCE)/365 "jours" 
FROM EMPLOYES 
FETCH NEXT 5 ROWS ONLY
```

# Tri du résultat d'une requête

>> **SELECT [ALL | DISTINCT]{*,[EXPRESSION1 [AS] ALIAS1[,...]}<br>
>> FROM NOM_TABLE<br>
>> ORDER BY <br>
>> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;[NOM_COLONNE1|ALIAS1|POSITION1|EXPRESSION1][ASC|DESC] [NULLS {FIRST|LAST}],<br>
>> &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;[NOM_COLONNE2|ALIAS2|POSITION2|EXPRESSION2][ASC|DESC] [NULLS {FIRST|LAST}]**<br>


```python
%%sql 
SELECT NOM, PRENOM, FONCTION 
FROM EMPLOYES 
ORDER BY NOM
```

```python
%%sql 
SELECT NOM||' '||PRENOM "Employé", FONCTION 
FROM EMPLOYES 
ORDER BY "Employé" ASC
```

Le tri se fait d'abord selon le premier critère spécifié dans la clause « **ORDER BY** », puis les lignes ayant la même valeur pour le premier critère sont triées selon le deuxième critère de la clause « **ORDER BY** », etc.

```python
%%sql 
SELECT NOM_PRODUIT, NO_FOURNISSEUR, CODE_CATEGORIE 
FROM PRODUITS
ORDER BY NO_FOURNISSEUR, CODE_CATEGORIE DESC
```

Si un attribut sur lequel porte un critère de tri contient la valeur « **NULL** », les lignes correspondantes sont affichées en dernier. Oracle traite les valeurs « **NULL** » comme si elles étaient des valeurs infinies.

```python
%%sql 
SELECT NOM,PRENOM,SALAIRE,COMMISSION
FROM EMPLOYES
ORDER BY COMMISSION
```

```python
%%sql 
SELECT NOM,PRENOM,SALAIRE,COMMISSION
FROM EMPLOYES
ORDER BY COMMISSION NULLS FIRST
```
